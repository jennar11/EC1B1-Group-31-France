```python
import pandas as pd
import numpy as np
import datetime as dt

## loading of Dataset for France. Variables for France have been merged together in 1 spreadsheeet 
france_data = pd.read_excel('France_Data.xlsx', index_col=0, skiprows=1)
france_data.index = pd.to_datetime(france_data.index)
display(france_data)
```

```python
## Loading of Dataset for US. Variables for US have been merged into 1 spreadsheet. 
us_data = pd.read_excel('US_Data.xlsx', index_col=0, skiprows=1)
us_data.index = pd.to_datetime(us_data.index)
display(us_data)
```

```python
## Merging Data for both France and US into 1 table 
merged_data = france_data.join(us_data,lsuffix='_France',rsuffix='_US')
merged_data.columns = ['Industrial Production (France)', 'Exchange Rate (France)', 'International Reserves and Liquidity (France)', 'Consumer Prices (France)','International Reserves and Liquidity (US)', 'Consumer Prices (US)']
display(merged_data)
```

```python
## France
## Log changes are used to accurately reflect percentage changes, as changes tend to be small (below 10 percent)

## Log Exchange Rate 
exchange_rate_france = merged_data['Exchange Rate (France)']
france_log_exchange_rate = np.log(exchange_rate_france) 

## Difference in Log Exchange Rate vs the Previous Month
france_difference_log_exchange_rate = france_log_exchange_rate.diff(1)

## Monthly Inflation Rate 
consumer_prices_france = merged_data['Consumer Prices (France)']
france_log_consumer_prices = np.log(consumer_prices_france)
france_monthly_inflation_rate = france_log_consumer_prices.diff(1)

## Log Industrial Production 
industrial_production_france = merged_data['Industrial Production (France)']
france_log_industrial_production = np.log(industrial_production_france)

## Monthly Growth in Industrial Production 
france_monthly_growth_industrial_production = france_log_industrial_production.diff(1)

## Growth in Industrial Production vs 12 Months Ago 
france_yearly_growth_industrial_production = france_log_industrial_production.diff(12)

## Index of the Value of International Reserves
int_reserves_france = merged_data['International Reserves and Liquidity (France)']
france_index_international_reserves = int_reserves_france/int_reserves_france[0]* 100

## Consolidate Variables and Present them in a Table
france_variables = pd.concat([france_log_exchange_rate, france_difference_log_exchange_rate, france_monthly_inflation_rate,
                              france_log_industrial_production, france_monthly_growth_industrial_production,
                              france_yearly_growth_industrial_production, france_index_international_reserves], axis=1)
france_variables.columns = ['France Log Exchange Rate', 'France Difference Log Exchange Rate', 'France Monthly Inflation Rate',
                           'France Log Industrial Production', 'France Monthly Growth Industrial Production', 
                            'France Yearly Growth Industrial Production', 'France Index International Reserves']
display(france_variables)
```

```python
## US

## Monthly Inflation Rate
us_consumer_prices = merged_data['Consumer Prices (US)']
us_log_consumer_prices = np.log(us_consumer_prices)
us_monthly_inflation_rate = us_log_consumer_prices.diff(1)

## Index of the Value of International Reserves
us_int_reserves = merged_data['International Reserves and Liquidity (US)']
us_index_international_reserves =us_int_reserves/us_int_reserves[0] * 100

## Consolidate Variables and Present them in a Table
us_variables = pd.concat([us_monthly_inflation_rate, us_index_international_reserves], axis=1)
us_variables.columns = ['US Monthly Inflation Rate', 'US Index International Reserves']
display(us_variables)
```

```python
## Combine the computed variables for France and US into 1 table 
combined_variables_list = [france_variables, us_variables]
combined_variables = pd.concat(combined_variables_list, axis=1)
display(combined_variables)
```

```python
## Inter-quartile range (IQR) proximity rule is used to identify and replace outliers.
## IQR is defined as the range from Quartile 1 - Quartile 3, the data between the 25th and 75th percentile of the dataset
## This rule defines outliers as values which lie outside the range of (q1 - 1.5 * IQR, q3 + 1.5 * IQR)

## Convert outliers into null values, to allow for easier interpolation later 
quartile_1 = combined_variables.quantile(0.25)
quartile_3 = combined_variables.quantile(0.75)
inter_quartile_range = quartile_3 - quartile_1
rejection_region = (combined_variables<(quartile_1 - 1.5 * inter_quartile_range)) | (combined_variables>(quartile_3 + 1.5 * inter_quartile_range))
combined_variables[rejection_region] = None

display(combined_variables)
```

```python
## Use linear interpolation to replace outliers 
## Linear interpolation is fast and often accurate enough, if the values in the data are closely spaced. 
combined_variables = combined_variables.interpolate(method ='linear', limit_direction = 'forward')
display(combined_variables)
```

```python
## Calculation on monthly real exchange rate growth 
## Change in Real Exchange Rate = Change in Nominal Exchange Rate + Domestic Inflation Rate - Foreign Inflation Rate 
france_monthly_real_exchange_rate_growth = combined_variables['France Difference Log Exchange Rate'] + combined_variables['France Monthly Inflation Rate'] - combined_variables['US Monthly Inflation Rate']
print(france_monthly_real_exchange_rate_growth)
```

```python
## Add monthly real exchange rate growth variable to existing table 
france_monthly_real_exchange_rate_growth = france_monthly_real_exchange_rate_growth.rename('Monthly Real Exchange Rate Growth')
combined_variables_list = [combined_variables, france_monthly_real_exchange_rate_growth]
combined_variables = pd.concat(combined_variables_list, axis=1)
display(combined_variables)
```

```python
## Calculation of difference in inflation between France and US
combined_variables['Difference in Inflation'] = combined_variables['France Monthly Inflation Rate'] - combined_variables['US Monthly Inflation Rate'] 

## Computation of data before and after exiting Bretton Woods 
before = combined_variables.index < dt.datetime(1970,10,1)
after = combined_variables.index > dt.datetime(1972,6,1)
combined_variables_before = combined_variables[before]
combined_variables_after = combined_variables[after]

print("Before exiting Bretton Woods")
before_standard_deviation = combined_variables_before.std(axis=0)
display(before_standard_deviation)

print("After exiting Bretton Woods")
after_standard_deviation = combined_variables_after.std(axis=0)
display(after_standard_deviation)
```

```python
## Ratio of standard deviations 
ratio = after_standard_deviation/before_standard_deviation

## Formulate a table listing the standard deviations of key variables (before leaving Bretton Woods, after leaving Bretton Woods, Ratio)
comparison_table_list = [before_standard_deviation, after_standard_deviation, ratio]
comparison_table = pd.concat(comparison_table_list, axis=1)
comparison_table.columns = ['Before Leaving Bretton Woods (SD)', 'After Leaving Bretton Woods (SD)', 'Ratio (SD)']
comparison_table_key_variables = comparison_table.iloc[[1,9,2,10,5],:]
display(comparison_table_key_variables)
```
